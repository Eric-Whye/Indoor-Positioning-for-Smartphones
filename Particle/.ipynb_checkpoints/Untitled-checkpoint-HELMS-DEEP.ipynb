```python
import pandas as pd
import numpy as np
from numpy import random
from numpy.random import uniform
import math
import plotly.express as px
```

```python
def create_uniform_particles(x_min, x_max, y_min, y_max, N):
    particles = np.empty((N, 2))
    particles[:, 0] = uniform(x_min, x_max, size=N)
    particles[:, 1] = uniform(y_min, y_max, size=N)
    return particles
        
#Apply movement to each particle
def apply_movement(particles, distance, angle):
    moved_particles = np.empty((len(particles), 2))
    for i in range(len(particles)):
        x = particles[i][0] + (distance * math.cos((angle * math.pi) / 180))
        y = particles[i][1] + (distance * math.sin((angle * math.pi) / 180))
        moved_particles[i] = [x,y]
    return moved_particles


def calculate_weights(particles, predicted_location):
    weights = []
    sum = 0
    #Get Sum of distances from all particles
    for particle in particles:
        d = distance(particle, predicted_location)
        sum += predict(d)
    for particle in particles:
        d = distance(particle, predicted_location)
        weight = predict(d) / sum
        weights.append(weight)

    #return likelihood for all particles between 0 and 1
    return weights

def distance(p1, p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)

a = 1
b = 2
def predict(distance):
    return np.exp(-(distance/a)**b)
    
    
def resample(particles, weights, N):
    new_particles = np.empty((N,2))
    
    for i in range(len(particles)):
        random.choice(np.arange(0,len(weights), 1), p=weights)
        
    #Return particles that tend towards the weights
```
